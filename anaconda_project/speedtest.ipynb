```python
 import numpy as np
 import time
 # Python list
 size = 1_000_000
 list1 = list(range(size))
 list2 = list(range(size))
 start = time.time()
 result = [x + y for x, y in zip(list1, list2)]
 end = time.time()
print("Python list addition time:", end - start)
 # NumPy array
 arr1 = np.array(list1)
 arr2 = np.array(list2)
 start = time.time()
 result = arr1 + arr2  # Vectorized operation
 end = time.time()
 print("NumPy array addition time:", end - start)
```

```python
import numpy as np
import time

size = 1_000_000   # big size for real comparison

list1 = list(range(size))

start = time.time()
list_squares = [x ** 2 for x in list1]
end = time.time()
print("Python list computation time:", end - start)

arr1 = np.array(list1)

start = time.time()
numpy_squares = arr1 ** 2
end = time.time()
print("NumPy array computation time:", end - start)
```

```python
arr = np.array([[10, 20, 30], [40, 50, 60]])
print("Shape:", arr.shape) # (2, 3) → 2 rows, 3 columns
print("Size:", arr.size)
print("Dimensions:", arr.ndim) # 2 → 2D array
# 6 → total elements
print("Data type:", arr.dtype) # int64 (or int32 on Windows)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.shape) # (2, 3)
reshaped = arr.reshape((3, 2)) # Change shape
print(reshaped)
 # [[1 2]
 #  [3 4]
 #  [5 6]]
flattened = arr.flatten() # Convert 2D → 1D
print(flattened) # [1 2 3 4 5 6]
```
